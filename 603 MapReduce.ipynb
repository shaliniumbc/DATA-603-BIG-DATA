```python
from functools import reduce

# Map function
def mapper(line):
    line = line.strip().lower()
    words = line.split(" ")
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    return word_count

# Reduce function
def reducer(word_counts1, word_counts2):
    for word, count in word_counts2.items():
        word_counts1[word] = word_counts1.get(word, 0) + count
    return word_counts1

with open("file1.txt", "r") as file:
    # Map step
    intermediate_output = map(mapper, file)

    # Reduce step
    Word_count = reduce(reducer, intermediate_output, {})

# Printing the word count
for word, count in Word_count.items():
    print(word, ":", count)
```

We are using a module in python called 'Functools' and importing function called reduce() and then we are performing map function and reduce function for the text file and then reading the input from the text file and then applying the mapper function for each line and emitting the intermediate key/value pairs adn in the reducer function we are combiniong the intermediate key/value pairs and calculating the final word/count and printing the number of words and how many times they are repeated.
